```python
import os
import sys

os.environ["PYSPARK_PYTHON"] = sys.executable
os.environ["PYSPARK_DRIVER_PYTHON"] = sys.executable

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master("local[1]") \
    .appName("RDDTest") \
    .config("spark.python.worker.reuse", "false") \
    .getOrCreate()

sc = spark.sparkContext
```

```python
pairRDD = sc.parallelize([
    ('HR','Ram'),
    ('HR','Riya'),
    ('IT','Amit')
])

pairRDD.groupByKey().collect()
```

```python
import sys
print(sys.executable)
```

```python
pairRDD.groupByKey() \
       .mapValues(list) \
       .collect()
```

```python
data = [
('A',10),
('B',20),
('C',30),
('B',40),
('A',50)
]

RDD1= sc.parallelize(data)
```

```python
RDD1.groupByKey().collect()
```

```python
RDD1.groupByKey() \
    .mapValues(list) \
    .collect()
```

```python
RDD1.groupByKey() \
    .mapValues(sum) \
    .collect()
```

```python
RDD1.reduceByKey(lambda x,y:x+y).collect()
```
