#### Find out the number of runs scored by each batsman

---

```python
#### WRITE YOUR CODE HERE
runs_per_batsman = df.groupBy("Batsman_Name").agg(F.sum("Runs").alias("Total_Runs"))
```

```python
# SHOW THE BATSMAN SCORED BY EACH RUN IN DESCENDING ORDER
runs_per_batsman.orderBy("Total_Runs",ascending=False).show()
```

---

#### Which batsman scored the highest number of boundaries

---

```python
## WRITE YOUR CODE HERE 
highest_no_of_boundaries = df.groupBy("Batsman_Name").agg(F.sum("Isboundary").alias("Total_Boundaries"))
```

```python
highest_no_of_boundaries.orderBy("Total_Boundaries",ascending=False).show(1)
```

**Define a `udf` function that will create a new column on the basis of following condition**

If the value of `Runs` is less than 2, then assign `A`, if value is between `3 to 5` then assign `B` else assign `C`


---

```python
## WRITE YOUR CODE HERE
from pyspark.sql.functions.builtin import when


s2=when(s1["Runs"]< 2, "A").when(s1["Runs"]< 4, "B").otherwise("C")
```

```python
s2=s1.withColumn("Runs_Category",s2).show()
```
